```python
import pandas as pd
```

```python
#Importing Excel File. Excel file contains 3 sheets. This creates a dictionary.
input_data = pd.read_excel('file.xlsx',sheet_name = None)
input_data
```

```python
input_data.keys()
```

```python
#concatenating all the 3 sheets into a dataframe
data=pd.concat(input_data.values(),ignore_index=False)
data
```

```python
#splitting the dataframe into sub-dataframes based on unique equipment code and put them into a dictionary.
dfs = {x: data[data['EQUIPMENT_CODE'] == x] for x in data['EQUIPMENT_CODE'].unique()}
dfs
```

```python
dfs.keys()
```

```python
final_df=pd.DataFrame()                         #Creating an empty dataframe in which updated dataframes will be appended.
for key in dfs.keys():                          #Loop through dictionary key items to get the unique equipment code
    df=dfs.get(key)                             #Each value of a dictionary is a dataframe
    df=df.apply(lambda x: x.dropna().reset_index(drop = True))                          #Loop through each column, remove duplicates,shifts the values up(by resetting index)
    df=df.dropna(subset=['PRIMARY_FUEL', 'SECONDARY_FUEL', 'CONTROL_TYPE'], how='all')  #subset without equipment code means - apart from equipment code, if mentioned column values are empty - drop the row.
    final_df=final_df.append(df)                                                        #store updated dataframes into final_df.
final_df=final_df.set_index('EQUIPMENT_CODE')                                           #update index as equipment code.
final_df.to_excel('final_df.xlsx')                                                      #Export final_df.
final_df
```
